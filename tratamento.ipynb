```python
import pyspark
import pandas as pd
from pyspark.sql.types import *
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, count, when, regexp_extract
```

```python
spark= SparkSession.builder.getOrCreate()
```

```python
# Leia o arquivo ‘videos-stats.csv'  no dataframe 'df_video' com cabeçalho e inferindo o esquema
df_video = spark.read.csv('videos-stats.csv', header=True, inferSchema=True)
df_video.show()
```

```python
# Altere os valores nulos dos campos 'Likes', 'Comments' e 'Views' para o valor 0
df_video = df_video.na.fill({'Likes': 0, 'Comments': 0, 'Views': 0})
df_video.show()
```

```python
# Leia o arquivo ‘comments.csv' no dataframe 'df_comentario' com cabeçalho e inferindo o esquema
df_comentario = spark.read.csv('comments.csv', header=True, inferSchema=True)
df_comentario.show()
```

```python
# Calcule a quantidade de registros do df_video e df_comentario
df_video = df_video.count()
df_comentario = df_comentario.count()
print(df_video)
print(df_comentario)
```

```python
schema_video2 = StructType([
    StructField('_c0', IntegerType(), True),
    StructField('Title', StringType(), True),
    StructField('Video ID', StringType(), True),
    StructField('Published At', StringType(), True),
    StructField('Keyword', StringType(), True),
    StructField('Likes', FloatType(), True),
    StructField('Comments', FloatType(), True),
    StructField('Views', FloatType(), True)
])

df_video2 = spark.read.csv('videos-stats.csv', header=True, schema=schema_video2)
df_video2.show()
```

```python
schema_comentario2 = StructType([
    StructField('_c0', StringType(), True),
    StructField('Video ID', StringType(), True),
    StructField('Comment', StringType(), True),
    StructField('Likes', IntegerType(), True),
    StructField('Sentiment', IntegerType(), True)
])

df_comentario2 = spark.read.csv('comments.csv', header=True, schema=schema_comentario2)
df_comentario2.show()
```

```python
# Remova os registros do df_video e df_comentario quem possuem o campo 'Video ID' nulos e calcule novamente a quantidade de registros
df_video = df_video2.na.drop(subset=['Video ID'])
df_comentario = df_comentario2.na.drop(subset=['Video ID'])
df_video = df_video.count()
df_comentario = df_comentario2.count()
print(df_video)
print(df_comentario2)
```

```python
# Remova os registros apenas do df_video quem possuem o campo 'Video ID' duplicados
df_video = df_video2.dropDuplicates(['Video ID'])
df_video.show()
```

```python
# Converta os campos Likes, Comments e Views para 'int' no dataframe df_video
df_video = df_video.withColumn('Likes', df_video['Likes'].cast(IntegerType()))
df_video = df_video.withColumn('Comments', df_video['Comments'].cast(IntegerType()))
df_video
```

```python
# Converta os campos Likes e Sentiment para 'int' no dataframe df_comentario,
# além disso, altere o nome do campo Likes para 'Likes Comment'

schema_comentario3 = StructType([
    StructField('_c0', StringType(), True),
    StructField('Video ID', StringType(), True),
    StructField('Comment', StringType(), True),
    StructField('Likes Comment', IntegerType(), True),
    StructField('Sentiment', IntegerType(), True)
])

df_comentario3 = spark.read.csv('comments.csv', header=True, schema=schema_comentario3)
df_comentario3
```

```python
# Crie o campo 'Interaction' no dataframe df_video, com a soma dos campos Likes, Comments e Views
df_video = df_video2.withColumn("Interaction", col("Likes") + col("Comments") + col("Views"))
df_video.show()
```

```python
# Converta os campos 'Published At' para 'date' no dataframe df_video
df_video = df_video2.withColumn("Published At", col("Published At").cast("date"))
df_video.show()
```

```python
# Crie o campo 'Year' no dataframe df_video, extraindo apenas o ano do campo 'Published At'
df_video = df_video2.withColumn("Year", regexp_extract(col("Published At"), r'(\d{4})', 1))
df_video.show()
```

```python
# Mescle os dados df_comentario no dataframe df_video em relação ao campo Video ID e crie o dataframe df_join_video_comments
df_join_video_comments = df_video2.join(df_comentario2, on='Video ID', how='inner')
df_join_video_comments.show(5)
```

```python
# Leia o arquivo ‘USvideos.csv' no dataframe 'df_us_videos' com cabeçalho e inferindo o esquema
df_us_videos = spark.read.csv('USvideos.csv', header=True, inferSchema=True)
df_us_videos.show(5)
```

```python
from os import truncate
# Mescle os dados df_us_videos no dataframe df_video em relação ao campo Title e crie e visualize o dataframe df_join_video_usvideos
df_join_videos_usvideos = df_video2.join(df_us_videos, on='Title', how='inner')
df_join_videos_usvideos.show(n=5, truncate=False)
```

```python
# Verifique a quantidade de campos nulos em todos os campos do dataframe df_video
df_video.select([count(when(col(c).isNull(), c)).alias(c) for c in df_video.columns]).show()
```

```python
# Remova a coluna '_c0' e salve o dataframe df_video como 'videos-tratados-parquet' no formato parquet e adicione o cabeçalho nos dados
df_video = df_video.drop('_c0')
df_video.write.csv('videos-tratados-parquet', header=True)
```

```python
# Remova a coluna '_c0' e salve o dataframe df_join_video_comments como 'videos-comments-tratados-parquet' no formato parquet e adicione o cabeçalho nos dados
df_join_video_comments = df_join_video_comments.drop('_c0')
if 'comment_likes' in df_join_video_comments.columns:
    df_join_video_comments = df_join_video_comments.drop('comment_likes')
elif 'Comment_Likes' in df_join_video_comments.columns:
    df_join_video_comments = df_join_video_comments.drop('Comment_Likes')
df_join_video_comments = df_join_video_comments.withColumnRenamed("Likes", "Comment_Likes")
df_join_video_comments.write.csv('videos-comments-tratados-parquet', header=True)
```
